# Everything together

```python
product_name = []
product_price = []
product_old_price = []
product_discount = []
product_cost_savings = []
product_qty_sold = []
product_offer_period_duration = []
product_url = []


for result in konga_sales:
    
    # Product name
    try:
        product_name.append(result.find('h3',{'class':'ec84d_3T7LJ'}).get_text())
    except:
        product_name.append('n/a')
        
    # product price
    try:
        product_price.append(result.find('span',{'class':'d7c0f_sJAqi'}).get_text())
    except:
        product_price.append('n/a')
    
    # product old price
    try:
        product_old_price.append(result.find('span',{'class':'f6eb3_1MyTu'}).get_text())
    except:
        product_old_price.append('n/a')
        
    # product discount
    try:
        product_discount.append(result.find('span',{'class':'_4472a_zYlL- _6c244_q2qap'}).get_text())
    except: 
        product_discount.append('n/a')
        
    # Product cost savings 
    try:
        product_cost_savings.append(result.find('span',{'class':'_279f7_29Idt'}).get_text())
    except:
        product_cost_savings.append('n/a')
    
    # product_qty_sold
    try:
        product_qty_sold.append(result.find('p',{'class':'f7113_1PcVu'}).get_text())
    except:
        product_qty_sold.append('n/a')
    
    # product_offer_period_duration
    try:
        product_offer_period_duration.append(result.find('time',{'class':'_01228_3ZMGW'}).get_text())
    except:
        product_offer_period_duration.append('n/a')
    
    #product_url
    try:
        product_url.append(result.find('a',{'class':'a2cf5_2S5q5 _2b5c5_1Qou0'}).get('href'))
    except:
        product_url.append('n/a')
        
    
```

## Combine urls

```python
url_combined = []

for link in product_url:
    url_combined.append(urllib.parse.urljoin(root_url,link))
```

```python
#url_combined
```

## Pandas Dataframe

```python
konga_practice = pd.DataFrame({'Name': product_name,
                              'Price': product_price,
                              'old price': product_old_price,
                              'discount': product_discount,
                              'quantity sold': product_qty_sold,
                              'cost savings': product_cost_savings,
                              'Duration for sale': product_offer_period_duration,
                              'Url': url_combined})
```

# Scraping Multiple Pages

```python
product_name = []
product_price = []
product_old_price = []
product_vendor = []
product_review =[]
product_discount = []
product_url = []

for i in range(1,25):
    
    # website
    website = 'https://www.konga.com/category/phones-tablets-5294?page=' + str(i)
    
    # requests
    response = requests.get(website)
    
    # soup object
    soup = bsoup(response.content, 'html.parser')
    
    # results 
    results = soup.find_all('li', {'class': 'bbe45_3oExY _22339_3gQb9'})
    
    # Loop through results
    
    for result in results:
        
        # Product name
        try:
            product_name.append(result.find('div',{'class':'af885_1iPzH'}).get_text())
        except:
            product_name.append('n/a')
            

        # product price
        try:
            product_price.append(result.find('span',{'class':'d7c0f_sJAqi'}).get_text())
        except:
            product_price.append('n/a')
            

        # product old price
        try:
            product_old_price.append(result.find('span',{'class':'f6eb3_1MyTu'}).get_text())
        except:
            product_old_price.append('n/a')
            

        # product discount
        try:
            product_discount.append(result.find('span',{'class':'_4472a_zYlL- _6c244_q2qap'}).get_text())
        except: 
            product_discount.append('n/a')
            

        # Product vendor 
        try:
            product_vendor.append(result.find('a',{'href':'/merchant/148506'}).get_text())
        except:
            product_vendor.append('n/a')
            

        # product_review
        try:
            product_review.append(result.find('span',{'class':'eea9b_1Ma8-'}).get_text())
        except:
            product_review.append('n/a')
            

        #product_url
        try:
            product_url.append(result.find('a').get('href'))
        except:
            product_url.append('n/a')
```

```python
url_combined = []

for link in product_url:
    url_combined.append(urllib.parse.urljoin(root_url,link))
```

```python
konga_practice = pd.DataFrame({'Name': product_name,
                              'Price': product_price,
                              'old price': product_old_price,
                              'discount': product_discount,
                               'vendoer':product_vendor,
                               'review': product_review,
                              'Url': url_combined})
```
